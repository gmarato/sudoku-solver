```python
import numpy as np
```

```python
# sudoku grid 9x9 as list of lists
# grid [row][coln]
grid = [[5,3,0,0,7,0,0,0,0],
        [6,0,0,1,9,5,0,0,0],
        [0,9,8,0,0,0,0,6,0],
        [8,0,0,0,6,0,0,0,3],
        [4,0,0,8,0,3,0,0,1],
        [7,0,0,0,2,0,0,0,6],
        [0,6,0,0,0,0,2,8,0],
        [0,0,0,4,1,9,0,0,5],
        [0,0,0,0,8,0,0,0,9]]
```

```python
print(np.matrix(grid))
```

```python
# given row and coln check if n can be placed
def ispossible(y, x, n):
    global grid
    
    # check if n can be placed in coln 
    for i in range(0,9):
        if grid[y][i] == n:
            return False
    # check if n can be placed in row
    for i in range(0,9):
        if grid[i][x] == n:
            return False

    # check if n can be placed in square
    x0 = (x//3)*3
    y0 = (y//3)*3
    for i in range(0,3):
        for j in range(0,3):
            if grid[y0+i][x0+j] == n:
                return False
    
    # if all conditions above not met
    return True
```

```python
def solve():
    global grid
   
    for y in range(9):
        for x in range(9):
            if grid[y][x] == 0:
                for n in range(1,10):
                    if ispossible(y,x,n):
                        grid[y][x] = n
                        solve()
                        grid[y][x] = 0
                return
    print("\n")            
    print(np.matrix(grid))
    input("More?")      
```

```python
solve()
```
